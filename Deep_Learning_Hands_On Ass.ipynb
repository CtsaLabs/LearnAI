```python
#import necessary libraries
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns

#model preprocessing, for splitting data
from sklearn.model_selection import train_test_split

#scaling our data helps our model, we will be using min_max scaler
from sklearn.preprocessing import MinMaxScaler

#score model accuracy
from sklearn.metrics import accuracy_score

#show accuracy mapping
from sklearn.metrics import confusion_matrix
```

```python
#get our dataset
df=pd.read_csv('data.csv')

df.head()
```

```python
#get informationn about the data
df.info()
```

```python
#get statistical informationn about the data
df.describe()
```

```python
#drop unnecessary columns
df=df.drop(['id', 'Unnamed: 32'], axis =1)
df.head()
```

```python
#you can explore other feature engineering
```

```python
#checking distribution of our target categorical column 
df.diagnosis.value_counts()

#our data is fairly balanced
```

```python
#visually exploring our target column distribution
#sns.countplot(df.diagnosis) this generated a value error
sns.countplot(x='diagnosis',data=df)
```

```python
#encoding the target to numbers
to_replace={'M': 1, 'B' : 0}

df.diagnosis.replace(to_replace, inplace=True)
```

```python
df.head()
```

```python
#separating the features from the target
X= df.drop('diagnosis', axis=1)
y= df.diagnosis
```

```python
X.head()
```

```python
y
```

```python
#scaling our data helps our model, we will be using min_max scaler
#scaling our data helps our model, we will be using min_max scaler
#from sklearn.preprocessing import MinMaxScaler
scaler = MinMaxScaler()
X_scaled= scaler.fit_transform(X)
```

```python
X_scaled
```

```python
#from sklearn.model_selection import train_test_split

#data splitting into test and train data
X_train, X_test, y_train, y_test = train_test_split(X_scaled,y, test_size=0.2, random_state=42)
```

```python
X_train
```

```python
 from xgboost import XGBClassifier

 xgb = XGBClassifier(
    n_estimators=300,
    learning_rate=0.01,
    max_depth=5,
    min_child_weight=2,
    subsample=0.7,
    colsample_bytree=0.9,
    gamma=0.1,
    scale_pos_weight=4.6
 )

 xgb.fit(X_train, y_train)

 y_pred= xgb.predict(X_test)

accuracy = accuracy_score(y_test, y_pred)
accuracy
```

```python
cm = confusion_matrix(y_test, y_pred)
cm
```

```python
sns.heatmap(cm, annot=True)
```
